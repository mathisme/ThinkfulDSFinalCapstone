```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data/df.csv')
df.info()
```

```python
df.columns
```

dropping that unnamed column which is just the index

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

Filtering for only those that have total blood cholesterol counts

```python
df = df[df.LBXTC.notnull()].copy()
df.info()
```

Filtering for only those that fasted 9 hours or more

```python
df = df[df.PHAFSTHR >= 9].copy()
df.drop('PHAFSTHR',axis=1, inplace=True)
df.info()
```

Filtering for those that are 20 or older

```python
df = df[df.RIDAGEYR>=20]
df.info()
```

Replacing systolic and diastolic blood pressures with the median of the four readings.

```python
df['Systolic'] = df[['BPXSY1','BPXSY2','BPXSY3','BPXSY4']].mean(axis=1)
df["Diastolic"] = df[['BPXDI1','BPXDI2','BPXDI3','BPXDI4']].mean(axis=1)
df.drop(['BPXSY1','BPXSY2','BPXSY3','BPXSY4','BPXDI1','BPXDI2','BPXDI3','BPXDI4'],axis=1,inplace=True)
```

For those that have not been told they have high blood pressure, BPQ020=2, I'm replacing BPQ050A, now taking prescription medication, with 2, no.  
For those that refused or don't know if they have been told they have high blood pressure, I am replacing BPQ020 and BPQ050A with np.nan
For those that have not been told to take prescription medication for high blood pressure, BPQ040A=2, I'm replacing BPQ050A with 2, no, as this question was skipped.
And for those that refused or don't know, I'm replacing the 050A value with np.nan
Then I'm dropping the variable BPQ040A, have you ever been told to take prescription medication

```python
df.loc[df.BPQ020==2,'BPQ050A']=2
df.loc[df.BPQ020==7,'BPQ020']=np.nan
df.loc[df.BPQ020==9,'BPQ020']=np.nan
df.loc[df.BPQ020==7,'BPQ050A']=np.nan
df.loc[df.BPQ020==9,'BPQ050A']=np.nan
df.loc[df.BPQ040A==2,'BPQ050A']=2
df.loc[df.BPQ040A==7,'BPQ050A']=np.nan
df.loc[df.BPQ040A==9,'BPQ050A']=np.nan
df.drop('BPQ040A',axis=1,inplace=True)
df.info()
```
