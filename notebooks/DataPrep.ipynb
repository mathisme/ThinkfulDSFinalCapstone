```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data/df.csv')
df.info()
```

```python
df.columns
```

dropping that unnamed column which is just the index

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

Filtering for only those that have total blood cholesterol counts

```python
df = df[df.LBXTC.notnull()].copy()
df.info()
```

Filtering for only those that fasted 9 hours or more

```python
df = df[df.PHAFSTHR >= 9].copy()
df.drop('PHAFSTHR',axis=1, inplace=True)
df.info()
```

Filtering for those that are 20 or older

```python
df = df[df.RIDAGEYR>=20]
df.info()
```
